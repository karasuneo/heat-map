```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import cartopy.crs as ccrs

# 三次元データを作成
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = [3, 4, 5, 6, 7]

# 地図のプロット
fig = plt.figure()
ax = fig.add_subplot(111, projection=ccrs.PlateCarree())
scatter = ax.scatter(x, y, c=z, cmap='viridis', s=100, transform=ccrs.PlateCarree())

# カラーバーの追加
cbar = plt.colorbar(scatter)
cbar.set_label('Z軸')

# グラフの表示
plt.show()
```
