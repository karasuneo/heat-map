```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# ランダムな点を生成
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# figureを生成
fig = plt.figure()

# axをfigureに設定
ax = fig.add_subplot(1, 1, 1, projection="3d")

# axesにplot
ax.scatter(x, y, z, c="b")

# 表示する
plt.show()
```
