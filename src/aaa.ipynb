```python
# 
latitudes = np.random.uniform(35, 36, 100)
longitudes = np.random.uniform(139, 140, 100)
timestamps = np.random.uniform(0, 24, 100)

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# 
sc = ax.scatter(longitudes, latitudes, timestamps, c=timestamps, cmap="viridis")

# 
cbar = plt.colorbar(sc)
cbar.set_label("Timestamp")

# 
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_zlabel("Timestamp")

plt.show()
```
